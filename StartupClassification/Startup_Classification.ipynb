```python
def split_into_x_and_y(df):
    """
    Utility to split df into X and y
    """
    df = df.dropna(subset=['status','country_code','market'])
    df = df.drop_duplicates()
    X = df.loc[:, df.columns != 'status']
    y = df.loc[:, df.columns == 'status']
    return X, y

X, y = split_into_x_and_y(df_startup)
```

### 2.2 Create a hold-out set

```python
def get_train_test_split(X, y, df=None):
    """
    Utility to create a train test split with a fixed random seed.
    """
    
    if df is not None:
        X_df, y_df = split_into_x_and_y(df)
    else:
        X_df, y_df = X, y
    return train_test_split(X_df, y_df, test_size=0.2, random_state=42)

X_viz_df, y_viz_df = X.copy(deep=True), y.copy(deep=True)
X_viz, X_viz_test, y_viz, y_viz_test = get_train_test_split(X_viz_df, y_viz_df)
```

```python
def make_preprocess_pipeline(training_X, text = False, target = True, discretize_year = False, impute_cont_strategy = 'mean'):
    list_of_processors = []
    global ohe_vars
    if text == False:
        assert text == False and target == True, "if text = False then target = True, otherwise too many features from one-hot-encode"
        cat_vars = list(training_X.columns[training_X.dtypes == 'object'])
        cont_vars = list(training_X.columns[training_X.dtypes != 'object'])
        if discretize_year:
            year_vars = {'last_funding_year', 'first_funding_year','founded_year'}
            cont_vars =  set(cont_vars) - year_vars
            year_transformer = make_pipeline(
                                    SimpleImputer(strategy=impute_cont_strategy),
                                    KBinsDiscretizer(n_bins=10, encode='ordinal'),
                                    StandardScaler())
            list_of_processors.append((year_transformer, list(year_vars)))
            cont_transformer = make_pipeline(SimpleImputer(strategy=impute_cont_strategy), 
                                       MaxAbsScaler())
            list_of_processors.append((cont_transformer, list(cont_vars)))
            target_vars = {'market','country_code', 'category_list'}
            cat_vars = list(set(cat_vars)-target_vars)
            target_transformer = make_pipeline(TargetEncoder(handle_missing='value', handle_unknown='value'), 
                                               StandardScaler())
            list_of_processors.append((target_transformer, list(target_vars)))
            ohe_transformer = make_pipeline(SimpleImputer(strategy = 'most_frequent'), 
                                            OneHotEncoder(handle_unknown='ignore'), MaxAbsScaler())
            
            ohe_vars = list(cat_vars)
            list_of_processors.append((ohe_transformer, list(cat_vars)))
                
        else:
            cont_transformer = make_pipeline(SimpleImputer(strategy=impute_cont_strategy), 
                                       MaxAbsScaler())
            list_of_processors.append((cont_transformer, cont_vars))
            if target:
                target_vars = {'market','country_code', 'category_list'}
                cat_vars = list(set(cat_vars)-target_vars)
                target_transformer = make_pipeline(TargetEncoder(handle_missing='value', handle_unknown='value'), 
                                                   StandardScaler())
                list_of_processors.append((target_transformer, list(target_vars)))
                ohe_transformer = make_pipeline(SimpleImputer(strategy = 'most_frequent'), 
                                                OneHotEncoder(handle_unknown='ignore'), MaxAbsScaler())

                ohe_vars = list(cat_vars)
                list_of_processors.append((ohe_transformer, list(cat_vars)))  
            else:
                ohe_transformer = make_pipeline(SimpleImputer(strategy = 'most_frequent'), 
                                                OneHotEncoder(handle_unknown='ignore'), MaxAbsScaler())

                ohe_vars = cat_vars
                list_of_processors.append((ohe_transformer, cat_vars))
    else:
        cat_vars = list(training_X.columns[training_X.dtypes == 'object'])
        cont_vars = list(training_X.columns[training_X.dtypes != 'object'])
        text_vars = {'category_list'}
        text_transformer = make_pipeline(CountVectorizer(stop_words = 'english', ngram_range = (1,1)))
        list_of_processors.append((text_transformer, 'category_list'))
        cat_vars = list(set(cat_vars)-text_vars)
        if discretize_year:
            year_vars = {'last_funding_year', 'first_funding_year','founded_year'}
            cont_vars =  set(cont_vars) - year_vars
            year_transformer = make_pipeline(
                                    SimpleImputer(strategy=impute_cont_strategy),
                                    KBinsDiscretizer(n_bins=10, encode='ordinal'),
                                    StandardScaler())
            list_of_processors.append((year_transformer, list(year_vars)))

            cont_transformer = make_pipeline(SimpleImputer(strategy=impute_cont_strategy), 
                                       MaxAbsScaler())
            list_of_processors.append((cont_transformer, list(cont_vars)))
            if target:
                target_vars = {'market','country_code'}
                cat_vars = list(set(cat_vars)-target_vars)
                target_transformer = make_pipeline(TargetEncoder(handle_missing='value', handle_unknown='value'), 
                                                   StandardScaler())
                list_of_processors.append((target_transformer, list(target_vars)))
                ohe_transformer = make_pipeline(SimpleImputer(strategy = 'most_frequent'), 
                                                OneHotEncoder(handle_unknown='ignore'), MaxAbsScaler())

                ohe_vars = list(cat_vars)
                list_of_processors.append((ohe_transformer, list(cat_vars)))
                
            else:
                ohe_transformer = make_pipeline(SimpleImputer(strategy = 'most_frequent'), 
                                                OneHotEncoder(handle_unknown='ignore'), MaxAbsScaler())

                ohe_vars = cat_vars
                list_of_processors.append((ohe_transformer, cat_vars))
        else:
            cont_transformer = make_pipeline(SimpleImputer(strategy=impute_cont_strategy), 
                                       MaxAbsScaler())
            list_of_processors.append((cont_transformer, cont_vars))
            if target:
                target_vars = {'market','country_code'}
                cat_vars = list(set(cat_vars)-target_vars)
                target_transformer = make_pipeline(TargetEncoder(handle_missing='value', handle_unknown='value'), 
                                                   StandardScaler())
                list_of_processors.append((target_transformer, list(target_vars)))
                ohe_transformer = make_pipeline(SimpleImputer(strategy = 'most_frequent'), 
                                                OneHotEncoder(handle_unknown='ignore'), MaxAbsScaler())

                ohe_vars = list(cat_vars)
                list_of_processors.append((ohe_transformer, list(cat_vars)))
                
            else:
                ohe_transformer = make_pipeline(SimpleImputer(strategy = 'most_frequent'), 
                                                OneHotEncoder(handle_unknown='ignore'), MaxAbsScaler())

                ohe_vars = cat_vars
                list_of_processors.append((ohe_transformer, cat_vars))
                
    transformer = make_column_transformer(*list_of_processors, remainder='passthrough')
    
    return transformer
            
```

```python
def get_features_names(transformer):
    features_names = []
    for i, (pipe_name, pipe, feature_list) in enumerate(transformer.transformers_):
        if not 'onehotencoder' in pipe.named_steps.keys() and not 'countvectorizer' in pipe.named_steps.keys():
            features_names.extend(feature_list)
        elif not 'onehotencoder' in pipe.named_steps.keys() and 'countvectorizer' in pipe.named_steps.keys():
            text_features = transformer.transformers_[i][1].named_steps['countvectorizer'].get_feature_names()
            features_names.extend(text_features)
        else:
            ohe_features = transformer.transformers_[i][1].named_steps['onehotencoder'].get_feature_names(input_features=ohe_vars)
            features_names.extend(ohe_features)
            if 'countvectorizer' in pipe.named_steps.keys():
                text_features = transformer.transformers_[i][1].named_steps['countvectorizer'].get_feature_names()
    return features_names
```

```python
def make_base_transformer(training_X):
    return make_preprocess_pipeline(training_X, text = False, target = True, discretize_year = False, impute_cont_strategy = 'mean'), "base_transformer"
def make_no_text_discretize_transformer(training_X):
    return make_preprocess_pipeline(training_X, text = False, target = True, discretize_year = True, impute_cont_strategy = 'mean'), "no_text_discretize_transformer"
def make_no_text_discretize_median_transformer(training_X):
    return make_preprocess_pipeline(training_X, text = False, target = True, discretize_year = True, impute_cont_strategy = 'median'), "no_text_discretize_median_transformer"
def make_text_target_discretize_transformer(training_X):
    return make_preprocess_pipeline(training_X, text = True, target = True, discretize_year = True, impute_cont_strategy = 'mean'), "text_target_discretize_transformer"
```

# Fixing imbalanced problem

### (1) Random Undersampling -> throwing away data points

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import make_pipeline as make_imb_pipeline
```

```python
def undersampling_model(transformer_function, model, training_X, training_y, testing_X, testing_y, matrix_report = False):
    transformer,name = transformer_function(training_X)
    model_pipe = make_pipeline(transformer, model)
    undersample_model_pipe = make_imb_pipeline(RandomUnderSampler(), model_pipe)
    undersample_model_score = cross_val_score(undersample_model_pipe, training_X, training_y, cv = 5, scoring= 'f1_micro').mean()
    print("Undersampling "+ str(model)+" model on "+ name + " will give micro f1 score: " + str(round(undersample_model_score, 6)))
    if matrix_report:
        undersample_model_pipe.fit(training_X, training_y)
        y_pred = undersample_model_pipe.predict(testing_X)
        print("Confusion matrix of this model on this transformer is:")
        print(confusion_matrix(testing_y, y_pred))
        print("Classification report of this model on this transformer is:")
        print(classification_report(testing_y, y_pred))
    pass
```

### (2) Oversampling -> double data points (take long)

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
def oversampling_model(transformer_function, model, training_X, training_y, testing_X, testing_y, matrix_report = False):
    transformer,name = transformer_function(training_X)
    model_pipe = make_pipeline(transformer, model)
    oversample_model_pipe = make_imb_pipeline(RandomOverSampler(), model_pipe)
    oversample_model_score = cross_val_score(oversample_model_pipe, training_X, training_y, cv = 5, scoring= 'f1_micro').mean()
    print("Oversampling "+ str(model)+" model on "+ name + " will give micro f1 score: " + str(round(oversample_model_score, 6)))
    if matrix_report:
        oversample_model_pipe.fit(training_X, training_y)
        y_pred = oversample_model_pipe.predict(testing_X)
        print("Confusion matrix of this model on this transformer is:")
        print(confusion_matrix(testing_y, y_pred))
        print("Classification report of this model on this transformer is:")
        print(classification_report(testing_y, y_pred))
    pass
```

### (3) Resampling within an ensemble -> not as long as oversampling

```python
from imblearn.ensemble import BalancedBaggingClassifier
```

```python
def resampled_ensemble(transformer_function, model, training_X, training_y, testing_X, testing_y, matrix_report = False):
    resampled_model = BalancedBaggingClassifier(base_estimator=model,random_state=0)
    transformer, name = transformer_function(training_X)
    resampled_model_pipe = make_imb_pipeline(transformer, resampled_model)
    resampled_model_score = cross_val_score(resampled_model_pipe, training_X, training_y, cv = 5, scoring= 'f1_micro').mean()
    print("Resampled Ensemble "+ str(model)+" model on "+ name + " will give micro f1 score: " + str(round(resampled_model_score, 6)))
    if matrix_report:    
        resampled_model_pipe.fit(training_X, training_y)
        y_pred = resampled_model_pipe.predict(testing_X)
        print("Confusion matrix of this model on this transformer is:")
        print(confusion_matrix(testing_y, y_pred))
        print("Classification report of this model on this transformer is:")
        print(classification_report(testing_y, y_pred))
    pass
```

### (4) SMOTE (Synthetic Minority Oversampling Technique)

```python
from imblearn.over_sampling import SMOTE
```

```python
def smote(transformer_function, model, training_X, training_y, testing_X, testing_y, matrix_report = False):
    transformer, name = transformer_function(training_X)
    smote_model_pipe = make_imb_pipeline(transformer, SMOTE(), model)
    smote_model_score = cross_val_score(smote_model_pipe, training_X, training_y, cv = 5, scoring= 'f1_micro').mean()
    print("SMOTE "+ str(model)+" model on "+ name + " will give micro f1 score: " + str(round(smote_model_score, 6)))
    if matrix_report: 
        smote_model_pipe.fit(training_X, training_y)
        y_pred = smote_model_pipe.predict(testing_X)
        print("Confusion matrix of this model on this transformer is:")
        print(confusion_matrix(testing_y, y_pred))
        print("Classification report of this model on this transformer is:")
        print(classification_report(testing_y, y_pred))
    pass
```

```python
def validate_models_on_transformer(list_of_classifiers, list_of_transformers, training_X, training_y, testing_X, testing_y):
    for transformer, classifier in list(product(list_of_transformers,list_of_classifiers)):
        print("===================================================================>")
        if classifier == tree:
            pass
        else:
            undersampling_model(transformer, classifier, training_X, training_y, testing_X, testing_y)
            print("------")
        if classifier == tree:
            pass
        else:
            oversampling_model(transformer, classifier, training_X, training_y, testing_X, testing_y)
            print("------")
        if classifier == forest:
            pass
        else:
            resampled_ensemble(transformer, classifier, training_X, training_y, testing_X, testing_y)
            print("------")
        if classifier == tree:
            pass
        else:
            smote(transformer, classifier, training_X, training_y, testing_X, testing_y)
            print("------")
    pass
```
